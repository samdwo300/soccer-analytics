```python
# Import data manipulation and analysis libraries
import pandas as pd  # for working with tabular data
import numpy as np   # for numerical operations
import matplotlib.pyplot as plt  # for plotting
import seaborn as sns  # for advanced visualizations
import unicodedata  # for removing accents and normalizing text
from unidecode import unidecode

# Load standard performance statistics for Serie A players
df_std = pd.read_csv('seriea_2024_2025_stats/seriea_standard_stats.csv')

# Load shooting-specific statistics for Serie A players
df_shoot = pd.read_csv('seriea_2024_2025_stats/seriea_shooting_stats.csv')

```

```python
# Merge key shooting statistics into the main standard stats dataframe
df_merged = pd.merge(
    df_std,
    df_shoot[['Player', 'Squad', 'Standard_Sh', 'Standard_SoT', 'Standard_Gls', 'Standard_PK', 'Standard_PKatt']],
    on=['Player', 'Squad'],  # Merge on both player and team name
    how='inner'  # Keep only players present in both datasets
)

# Define a function to remove accents from text (e.g., "é" -> "e")
def remove_accents(text):
    if pd.isnull(text):
        return ""
    return unidecode(str(text))

# Clean and standardize player and squad names for consistency
df_merged['Player'] = df_merged['Player'].apply(remove_accents).str.lower().str.strip()
df_merged['Squad'] = df_merged['Squad'].apply(remove_accents).str.lower().str.strip()

# Convert relevant columns to numeric types for calculation
df_merged['Min'] = pd.to_numeric(df_merged['Playing Time_Min'], errors='coerce')  # Minutes played
df_merged['Goals'] = pd.to_numeric(df_merged['Performance_Gls'], errors='coerce')  # Goals scored
df_merged['Shots'] = pd.to_numeric(df_merged['Standard_Sh'], errors='coerce')  # Total shots
df_merged['SoT'] = pd.to_numeric(df_merged['Standard_SoT'], errors='coerce')  # Shots on target

# Calculate per-player shooting metrics
df_merged['Goals_per90'] = df_merged['Goals'] / (df_merged['Min'] / 90)  # Goals per 90 minutes
df_merged['Shot_Conversion'] = df_merged['Goals'] / df_merged['Shots']  # % of shots that became goals
df_merged['SoT_pct'] = df_merged['SoT'] / df_merged['Shots']  # % of shots that were on target
```

```python
# Create a working copy of the merged dataset focused on attacking players
attackers = df_merged.copy()

# Filter out players with low sample sizes to ensure meaningful analysis
# - At least 20 total shots
# - At least 5 goals
# - At least 10 matches played
attackers = attackers[
    (attackers['Shots'] >= 20) &
    (attackers['Goals'] >= 5) &
    (attackers['Playing Time_MP'] >= 10)
]
```

```python
# Calculate conversion rate and xG per shot
attackers['Shot_Conversion'] = attackers['Goals'] / attackers['Shots']
attackers['xG_per_shot'] = attackers['Expected_xG'] / attackers['Shots']
xg_mean = attackers['xG_per_shot'].mean()
conv_mean = attackers['Shot_Conversion'].mean()
```

```python
import pandas as pd

# Load the file normally
all_serie_salary = pd.read_csv('seriea_2024_2025_stats/all_serie-a_2024-2025.csv')

import pandas as pd


# Drop the first row if it contains duplicate headers
all_serie_salary = all_serie_salary.iloc[1:].reset_index(drop=True)

# Clean column names
all_serie_salary.columns = [col.split("  ")[0].strip() if isinstance(col, str) else col for col in all_serie_salary.columns]

# Rename if desired
# Check which column is the salary (index 4 in this case)
all_serie_salary.columns.values[1] = 'Weekly Salary'
all_serie_salary.columns.values[2] = 'Yearly Salary'
all_serie_salary.rename(columns={'Unnamed: 0': 'Player'}, inplace=True)
```

```python
# Remove any rows where 'Weekly Salary' equals the column name
all_serie_salary = all_serie_salary[all_serie_salary['Weekly Salary'] != 'Gross P/W (EUR)']
all_serie_salary = all_serie_salary[all_serie_salary['Yearly Salary'] != 'Gross P/Y (EUR)']


currency_columns = ['Weekly Salary', 'Yearly Salary']

for col in currency_columns:
    if col in all_serie_salary.columns:
        all_serie_salary[col] = (
            all_serie_salary[col]
            .astype(str)
            .str.replace('£', '', regex=False)
            .str.replace('€', '', regex=False)
            .str.replace(',', '', regex=False)
            .str.strip()
            .replace('', '0')  # optional: convert empty string to 0
            .astype(float)
        )
```

```python
all_serie_salary['Player'] = all_serie_salary['Player'].apply(remove_accents).str.lower().str.strip()
all_serie_salary['Team'] = all_serie_salary['Team'].apply(remove_accents).str.lower().str.strip()
```

```python
df_sal_stats = pd.merge(
    all_serie_salary,
    attackers[['Player', 'Squad', 'Standard_Sh', 'Standard_SoT', 'Standard_Gls', 'Standard_PK', 'Standard_PKatt','Expected_xG','xG_per_shot','Shot_Conversion']],
    on=['Player'],  # Merge on both player and team name
    how='right'  # Keep only players present in both datasets
)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(14, 9))

# Calculate G – xG
df_sal_stats['G_minus_xG'] = df_sal_stats['Standard_Gls'] - df_sal_stats['Expected_xG']

# --- Bubble size: scale yearly salary with square-root scaling ---
max_salary = df_sal_stats['Yearly Salary'].max()
salary_sizes = (
    np.sqrt(df_sal_stats['Yearly Salary'].fillna(0)) / np.sqrt(max_salary) * 1200
)

# --- Color based on G - xG ---
cmap = plt.get_cmap('RdYlGn')
norm = plt.Normalize(df_sal_stats['G_minus_xG'].min(), df_sal_stats['G_minus_xG'].max())

# --- Scatter plot ---
sc = plt.scatter(
    df_sal_stats['xG_per_shot'],
    df_sal_stats['Shot_Conversion'],
    s=salary_sizes,
    c=df_sal_stats['G_minus_xG'],
    cmap=cmap,
    norm=norm,
    alpha=0.85,
    edgecolor='k'
)

# --- Annotate standout players ---
# --- Annotate standout players and Lautaro ---
for _, row in df_sal_stats.iterrows():
    name = row['Player'].lower()
    if 'lautaro' in name and 'martinez' in name:
        plt.text(
            row['xG_per_shot'],
            row['Shot_Conversion'] + 0.01,
            row['Player'].title(),
            fontsize=8,
            ha='center'
        )
    elif (abs(row['G_minus_xG']) > 2 or row['Shot_Conversion'] > df_sal_stats['Shot_Conversion'].mean() + 0.05) and row['Standard_Sh'] > 15:
        plt.text(
            row['xG_per_shot'],
            row['Shot_Conversion'] + 0.01,
            row['Player'].title(),
            fontsize=8,
            ha='center'
        )

# --- Axes and reference lines ---
plt.axvline(df_sal_stats['xG_per_shot'].mean(), color='red', linestyle='--', linewidth=1, label='Avg xG/Shot')
plt.axhline(df_sal_stats['Shot_Conversion'].mean(), color='blue', linestyle='--', linewidth=1, label='Avg Shot Conversion')
plt.xlabel('xG per Shot (Chance Quality)')
plt.ylabel('Shot Conversion Rate (Goals/Shots)')
plt.title('Serie A: Finishing, Value-for-Money, and Shot Quality (2024–25, ≥20 shots)')

# --- Colorbar for G - xG ---
cbar = plt.colorbar(sc)
cbar.set_label('Goals minus xG (G - xG)')

# --- Dummy legend for salary bubble size ---
for s in [1, 3, 6]:  # Salaries in millions
    size = np.sqrt(s * 1_000_000) / np.sqrt(max_salary) * 1000
    plt.scatter([], [], s=size, c='gray', alpha=0.5, edgecolor='k', label=f'€{s}M Salary')

plt.legend(scatterpoints=1, frameon=True, labelspacing=1, title='Annual Salary', loc='lower right')

plt.grid(alpha=0.3)
plt.tight_layout()


# Save and show the final combined plot
plt.savefig("seriea_quadrant_dots_salary_g_xg_color.png", dpi=300, bbox_inches='tight')
plt.show()
```
